```python
#!pip install matplotlib
#!pip install seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
banco = pd.read_csv("bb.csv", parse_dates=['Data Finalização'])
```

```python
banco.info()
```

```python
banco['Nota do Consumidor'] = banco['Nota do Consumidor'].astype(int)
```

```python
banco['Tempo Resposta'] = banco['Tempo Resposta'].astype(int)
```

```python
banco.head()
```

```python
numericas = banco.describe().reset_index()
numericas.columns = ["medida", "Tempo Resposta", "Nota do Consumidor"]
numericas
```

```python
sns.countplot(y='Nota do Consumidor', data=banco)
```

```python
import numpy as np
banco['nota insatisfatória'] = np.where((banco["Nota do Consumidor"]==1) |(banco["Nota do Consumidor"]==2),1,0)
```

```python
banco['nota insatisfatória'].value_counts(1)
```

```python
sns.countplot(y='nota insatisfatória', data=banco)
```

```python
categoricas = banco.describe(include=['O']).reset_index()
```

```python
regiao = banco.Região.value_counts(1).reset_index()
regiao.columns = ['Região', 'Frequência']
regiao
```

```python
uf = banco.UF.value_counts(1).reset_index()
uf.columns = ['UF', 'Frequência']
uf
```

```python
cidade = banco.Cidade.value_counts(1).reset_index()
cidade.columns = ['Cidade', 'Frequência']
cidade
```

```python
sexo = banco.Sexo.value_counts(1).reset_index()
sexo.columns = ['Sexo', 'Frequência']
sexo
```

```python
faixa_etaria = banco['Faixa Etária'].value_counts(1).reset_index()
faixa_etaria.columns = ["Faixa Etária", "Frequência"]
faixa_etaria
```

```python
data_finalizacao = banco["Data Finalização"].dt.year.value_counts()
data_finalizacao
```

```python
assunto = banco.Assunto.value_counts(1).reset_index()
assunto.columns = ["Assunto", "Frequência"]
assunto
```

```python
g_problema = banco["Grupo Problema"].value_counts(1).reset_index()
g_problema.columns = ["Grupo Problema", "Frequência"]
g_problema
```

```python
problema = banco.Problema.value_counts(1).reset_index()
problema.columns = ["Problema", "Frequência"]
problema
```

```python
como_comprou = banco["Como Comprou Contratou"].value_counts(1).reset_index()
como_comprou.columns = ["Como Comprou Contratou", "Frequência"]
como_comprou
```

```python
procurou_empresa = banco["Procurou Empresa"].value_counts(1).reset_index()
procurou_empresa.columns = ["Procurou Empresa", "Frequência"]
procurou_empresa
```

```python
respondida = banco.Respondida.value_counts(1).reset_index()
respondida.columns = ["Respondida", "Frequência"]
respondida
```

```python
situacao = banco.Situação.value_counts(1).reset_index()
situacao.columns = ["Situação", "Frequência"]
situacao
```

```python
avaliacao = banco["Avaliação Reclamação"].value_counts(1).reset_index()
avaliacao.columns = ["Avaliação Reclamação", "Frequência"]
avaliacao
```

```python
with pd.ExcelWriter('tabelas.xlsx') as writer1:
    numericas.to_excel(writer1, sheet_name = 'numericas', index = False)
    categoricas.to_excel(writer1, sheet_name = 'categoricas', index = False)
    regiao.to_excel(writer1, sheet_name = 'regiao', index = False)
    uf.to_excel(writer1, sheet_name = 'uf', index = False)
    cidade.to_excel(writer1, sheet_name = 'cidade', index = False)
    sexo.to_excel(writer1, sheet_name = 'sexo', index = False)
    faixa_etaria.to_excel(writer1, sheet_name = 'faixa_etaria', index = False)
    data_finalizacao.to_excel(writer1, sheet_name = 'data_finalizacao', index = False)
    assunto.to_excel(writer1, sheet_name = 'assunto', index = False)
    g_problema.to_excel(writer1, sheet_name = 'g_problema', index = False)
    problema.to_excel(writer1, sheet_name = 'problema', index = False)
    como_comprou.to_excel(writer1, sheet_name = 'como_comprou', index = False)
    procurou_empresa.to_excel(writer1, sheet_name = 'procurou_empresa', index = False)
    respondida.to_excel(writer1, sheet_name = 'respondida', index = False)
    situacao.to_excel(writer1, sheet_name = 'situacao', index = False)
    avaliacao.to_excel(writer1, sheet_name = 'avaliacao', index = False)
    
    
```
